## Fetch Dataset from Google Drive

Copies `train.7z` and `val.7z` from Google Drive and extracts them into the expected COCO layout:
```
data/coco/
  train/
    coco_instances.json
    images/
    masks/      (optional)
  val/
    coco_instances.json
    images/
    masks/      (optional)
```

````python
import os
import shutil
import subprocess

# Clean up any previous local data
if os.path.exists(DATA_LOCAL_PATH):
    shutil.rmtree(DATA_LOCAL_PATH)
os.makedirs(DATA_LOCAL_PATH, exist_ok=True)

def copy_and_extract(split: str):
    archive_name = f'{split}.7z'
    gdrive_archive = os.path.join(GDRIVE_DATASET_PATH, archive_name)
    local_archive  = os.path.join(DATA_LOCAL_PATH, archive_name)
    dest_dir       = os.path.join(DATA_LOCAL_PATH, split)

    if not os.path.isfile(gdrive_archive):
        raise FileNotFoundError(
            f"Expected archive not found on Google Drive:\n  {gdrive_archive}\n"
            f"Please upload {archive_name} to: {GDRIVE_DATASET_PATH}"
        )

    print(f'Copying {archive_name} from Google Drive ...')
    shutil.copy(gdrive_archive, local_archive)
    print(f'✓ Copied ({os.path.getsize(local_archive) / 1024**2:.1f} MB)')

    print(f'Extracting {archive_name} ...')
    os.makedirs(dest_dir, exist_ok=True)
    result = subprocess.run(
        ['7z', 'x', local_archive, f'-o{dest_dir}', '-y'],
        capture_output=True, text=True
    )
    if result.returncode != 0:
        raise RuntimeError(f'Extraction failed:\n{result.stderr}')

    os.remove(local_archive)

    # Verify expected files
    ann_file  = os.path.join(dest_dir, 'coco_instances.json')
    img_dir   = os.path.join(dest_dir, 'images')
    n_images  = len(os.listdir(img_dir)) if os.path.isdir(img_dir) else 0
    ann_ok    = os.path.isfile(ann_file)
    print(f'✓ {split}: {n_images} images, annotation file {"found" if ann_ok else "MISSING"}')

    if not ann_ok:
        raise FileNotFoundError(
            f'coco_instances.json not found in {dest_dir}.\n'
            'Make sure the archive extracts to: {split}/coco_instances.json'
        )

print('=== Fetching training data ===')
copy_and_extract('train')

print('\n=== Fetching validation data ===')
copy_and_extract('val')

# Read and print dataset info
import json
with open(os.path.join(DATA_LOCAL_PATH, 'train', 'coco_instances.json')) as f:
    ann = json.load(f)

cats = ann['categories']
num_classes     = max(c['id'] for c in cats) + 1
num_body_points = max(len(c['keypoints']) for c in cats)

print(f'\n=== Dataset summary ===')
print(f'  Categories       : {len(cats)}')
print(f'  Max category ID  : {max(c["id"] for c in cats)}  →  num_classes = {num_classes}')
print(f'  Keypoints/instance: {num_body_points}')
print(f'  Training images  : {len(ann["images"])}')
print(f'  Training annots  : {len(ann["annotations"])}')
````

## Train Model

Runs `train.py` with the selected custom config.  
Output is streamed live and periodically backed up to Google Drive every `BACKUP_EVERY_N_EPOCHS` epochs.

````python

import datetime
import os
import shutil
import subprocess
from zoneinfo import ZoneInfo

python_path = os.path.join(VENV_PATH, 'bin', 'python')

# ── Build training command ────────────────────────────────────────────────────
cmd = [python_path, 'train.py', '--config_file', CONFIG_FILE]

# String values must be quoted so ast.literal_eval treats them as str, not identifiers
options = [f'training_params.output_dir="{OUTPUT_DIR}"']

# Pass backup configuration to trainer
options.append(f'training_params.backup_every_n_epochs={BACKUP_EVERY_N_EPOCHS}')
options.append(f'training_params.gdrive_backup_path="{GDRIVE_BACKUP_PATH}"')

if EPOCHS is not None:
    options.append(f'training_params.epochs={EPOCHS}')
    # Scale LR decay milestone to 80% of total epochs
    lr_milestone = int(EPOCHS * 0.8)
    options.append(f'lr_scheduler.milestones=[{lr_milestone}]')

if BATCH_SIZE is not None:
    options.append(f'dataset_train.total_batch_size={BATCH_SIZE}')
    options.append(f'dataset_val.total_batch_size={BATCH_SIZE}')
    options.append(f'dataset_test.total_batch_size={BATCH_SIZE}')

if NUM_WORKERS is not None:
    options.append(f'dataset_train.num_workers={NUM_WORKERS}')
    options.append(f'dataset_val.num_workers={NUM_WORKERS}')
    options.append(f'dataset_test.num_workers={NUM_WORKERS}')

cmd += ['--options'] + options

if USE_AMP:
    cmd.append('--amp')

if EVAL_INTERVAL is not None and EVAL_INTERVAL > 1:
    cmd += ['--eval_interval', str(EVAL_INTERVAL)]

if RESUME_TRAINING:
    resume_ckpt = os.path.join(REPO_LOCAL_PATH, OUTPUT_DIR, RESUME_CHECKPOINT)
    if not os.path.isfile(resume_ckpt):
        # Try to copy latest checkpoint from Google Drive backup
        gdrive_ckpt = os.path.join(GDRIVE_BACKUP_PATH, RESUME_CHECKPOINT)
        if os.path.isfile(gdrive_ckpt):
            os.makedirs(os.path.dirname(resume_ckpt), exist_ok=True)
            shutil.copy(gdrive_ckpt, resume_ckpt)
            print(f'✓ Checkpoint copied from Google Drive backup')
        else:
            raise FileNotFoundError(
                f'Resume checkpoint not found locally or on Google Drive:\n'
                f'  Local   : {resume_ckpt}\n'
                f'  GDrive  : {gdrive_ckpt}'
            )
    cmd += ['--resume', resume_ckpt]
    print(f'Resuming from: {resume_ckpt}')

if PRETRAIN_LOCAL_PATH:
    cmd += ['--pretrain', PRETRAIN_LOCAL_PATH]

# ── Environment ───────────────────────────────────────────────────────────────
train_env = os.environ.copy()
train_env['CUDA_HOME']        = '/content/venv/cuda'
train_env['PATH']             = f"/content/venv/cuda/bin:{train_env.get('PATH', '')}"
train_env['LD_LIBRARY_PATH']  = f"/content/venv/cuda/lib64:{train_env.get('LD_LIBRARY_PATH', '')}"
train_env['CUDNN_PATH']       = '/content/venv/cuda'
train_env['MPLBACKEND']       = 'Agg'

# ── Run training ─────────────────────────────────────────────────────────────
tz = ZoneInfo('Africa/Johannesburg')
start_dt = datetime.datetime.now(datetime.timezone.utc).astimezone(tz)
print(f'Start time (+02:00): {start_dt.strftime("%Y/%m/%d %H:%M:%S")}')
print(f'Command: {" ".join(cmd)}\n')
print('=' * 80)

process = subprocess.Popen(
    cmd,
    cwd=REPO_LOCAL_PATH,
    env=train_env,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1,
    universal_newlines=True
)

for line in process.stdout:
    print(line, end='', flush=True)

process.wait()
returncode = process.returncode

print('=' * 80)
end_dt = datetime.datetime.now(datetime.timezone.utc).astimezone(tz)
print(f'End time (+02:00): {end_dt.strftime("%Y/%m/%d %H:%M:%S")}')
elapsed = end_dt - start_dt
print(f'Elapsed: {str(elapsed).split(".")[0]}')

if returncode != 0:
    print(f'⚠ Training failed with exit code {returncode}')
else:
    print('✓ Training completed successfully')

# ── Copy Results to Google Drive ─────────────────────────────────────────────
now_dt = datetime.datetime.now(datetime.timezone.utc).astimezone(tz)
date_str = now_dt.strftime('%Y_%m_%d_h%H_m%M_s%S')
run_label = f'{date_str}_{MODEL_SIZE}_ep{EPOCHS}'

output_local  = os.path.join(REPO_LOCAL_PATH, OUTPUT_DIR)
output_gdrive = os.path.join(GDRIVE_OUTPUT_PATH, f'runs_{run_label}')

# Only copy results if training succeeded and output exists
if returncode == 0 and os.path.isdir(output_local):
    print(f'\n=== Copying training results to Google Drive ===')
    os.makedirs(GDRIVE_OUTPUT_PATH, exist_ok=True)
    shutil.copytree(output_local, output_gdrive)
    print(f'✓ Results saved to: {output_gdrive}')

    # Clean up the temporary training backup folder
    if os.path.isdir(GDRIVE_BACKUP_PATH):
        print(f'\n=== Cleaning up temporary training backup ===')
        shutil.rmtree(GDRIVE_BACKUP_PATH)
        print(f'✓ Removed: {GDRIVE_BACKUP_PATH}')

    # Print checkpoint summary
    ckpts = [f for f in os.listdir(output_gdrive) if f.endswith('.pth')]
    print(f'\n=== Training Summary ===')
    print(f'Checkpoints saved ({len(ckpts)}): {", ".join(sorted(ckpts))}')
    print(f'Final results location: {output_gdrive}')
elif returncode != 0:
    print('\n⚠ Skipping results copy due to training failure')
else:
    print(f'\n⚠ Output directory not found: {output_local}')
````

## (Dev Only) Archive Packages for Cached Retrieval

Run this once after a successful fresh install to create cached archives on Google Drive,  
dramatically speeding up environment setup on future Colab runs.

````python
import os
import shutil
import subprocess

os.makedirs(GDRIVE_INSTALLER_PATH, exist_ok=True)

def _archive_to_gdrive(src_dir, archive_name, cwd, gdrive_dest):
    """Create a 7z archive of src_dir and copy it to Google Drive."""
    if os.path.exists(gdrive_dest):
        print(f'Archive already exists, skipping: {archive_name}')
        return
    local_archive = f'/content/{archive_name}'
    if os.path.exists(local_archive):
        os.remove(local_archive)
    print(f'Creating {archive_name} ...')
    subprocess.run(
        ['7z', 'a', '-mx=9', local_archive, os.path.basename(src_dir)],
        cwd=cwd, check=True, capture_output=True
    )
    size_mb = os.path.getsize(local_archive) / 1024**2
    print(f'  Archive size: {size_mb:.1f} MB')
    shutil.copy(local_archive, gdrive_dest)
    os.remove(local_archive)
    print(f'\u2713 Saved to Google Drive: {gdrive_dest}')

# Archive CUDA/cuDNN
print('=== Archiving CUDA/cuDNN ===')
_archive_to_gdrive(
    src_dir='/content/venv/cuda',
    archive_name=CACHED_CUDA_ARCHIVE,
    cwd='/content/venv',
    gdrive_dest=CACHED_CUDA_PATH
)

# Archive site-packages
print('\n=== Archiving site-packages ===')
_archive_to_gdrive(
    src_dir=os.path.join(VENV_PATH, 'lib', 'python3.10', 'site-packages'),
    archive_name=CACHED_SITE_PACKAGES_ARCHIVE,
    cwd=os.path.join(VENV_PATH, 'lib', 'python3.10'),
    gdrive_dest=CACHED_SITE_PACKAGES_PATH
)

print(f'\n=== Done ===')
print(f'Archives location: {GDRIVE_INSTALLER_PATH}')
print('Delete the archives from Google Drive to regenerate them.')
````
